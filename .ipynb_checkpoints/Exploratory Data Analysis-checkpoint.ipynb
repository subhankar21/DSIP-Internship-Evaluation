```python
import sys
import warnings
warnings.filterwarnings("ignore")
sys.path.append('utils')
import pandas as pd
from loader_utils import url_dframe
from EDA_utils import *
```

```python
weblink = "https://github.com/CapitalOneRecruiting/DS/raw/master/transactions.zip"
cap_df = url_dframe(weblink)
cap_df.head()
```

```python
print(f"Number of records: {cap_df.shape[0]}")
print(f"Number of fields: {cap_df.shape[1]}")
```

```python
# The basic overall summary of all the field.
summary_statistics(cap_df)
```

```python
cap_df['year'] = pd.to_datetime(cap_df['accountOpenDate'], errors='coerce').dt.year
yearly_counts = cap_df['year'].value_counts().sort_index()
line_plot(yearly_counts, 'Year','Accounts Opened', 'Account Growth Over Time')
```

```python
country_count = cap_df['acqCountry'].value_counts().sort_index()
print('The Distribution of Account Country Specific:',country_count)
```

```python
cap_df['acqCountry'], cap_df['availableMoney']
total_amount_by_country = cap_df.groupby('acqCountry')['availableMoney'].sum()
dollar_format(total_amount_by_country )
```

```python
print('There are total', len(cap_df['cardCVV'].value_counts().sort_index()), 'repetative CVV ')
```

```python
print('Total number of Bank Card issued')
print(cap_df['cardPresent'].value_counts().sort_index())
```

```python
print('Maximum Card Limit in dollar Issued:', cap_df['creditLimit'].max())
print('Minimum Card Limit in dollar Issued:',cap_df['creditLimit'].min())
print('Total count of Particular Card Limit in dollar Issued:', cap_df['creditLimit'].value_counts().sort_index())
```

```python
print(f"Count of credit balance less than $500: {len(cap_df[cap_df['currentBalance'] < 500])}")
print(f"Count of credit balance more than $500: {len(cap_df[cap_df['currentBalance'] >= 500])}")
min_balance = cap_df['currentBalance'].min()
print(f"Total of {len(cap_df[cap_df['currentBalance'] == min_balance])} customer bank card has ${min_balance} credit balance")
max_balance = cap_df['currentBalance'].max()
print(f"Total of {len(cap_df[cap_df['currentBalance'] == max_balance])} customer bank card has ${max_balance} credit balance")
```

```python
fraud_df = cap_df[cap_df['isFraud'] == True]
print(f"Total number of Fraud Account: {len(fraud_df )}")
maximum_fraud_cred = len(fraud_df[fraud_df['creditLimit'] == fraud_df['creditLimit'].max()])
minimum_fraud_cred = len(fraud_df[fraud_df['creditLimit'] == fraud_df['creditLimit'].min()])
print(f"Number of account with maximum credit limit of ${fraud_df['creditLimit'].max()} is Fraud: {maximum_fraud_cred}")
print(f"Number of account with minimum credit limit of ${fraud_df['creditLimit'].min()} is Fraud: {minimum_fraud_cred}")
maximum_balance = len(fraud_df[fraud_df['currentBalance'] == fraud_df['currentBalance'].max()])
minimum_balance = len(fraud_df[fraud_df['currentBalance'] == fraud_df['currentBalance'].min()])
print(f"Number of account with maximum account balance of ${fraud_df['currentBalance'].max()} is Fraud: {maximum_balance}")
print(f"Number of account with minimum account balance of ${fraud_df['currentBalance'].min()} is Fraud: {minimum_balance}")
```

```python
mer_code = cap_df['merchantCategoryCode'].value_counts().sort_index()
fig_size = (12,6)
yearly_bar_plot(mer_code , 'Merchant Category','Frequency', 'Distribution of Purchases Across Categories',fig_size )
```
